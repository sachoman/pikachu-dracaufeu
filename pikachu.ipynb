```python
import pefile
import base64

input_file_name = 'pikachu.exe'
pe = pefile.PE(input_file_name)

# Adresse de début de la zone qui nous intéresse dans IDA
string_address = 0x1400082E0

#Adresse de début de la section .rdata dans IDA
base_rdata_address = 0x140003000

# Calcul de l'offset des données dans la section .rdata
offset = string_address  - base_rdata_address

# Nombre d'octets à lire (19392)
bytes_to_read = 0x4bc0

# Parcours des sections pour trouver la section .rdata
for section in pe.sections:
    if '.rdata' in section.Name.decode().rstrip('\x00'):
        # Lecture des données 
        rdata_data = section.get_data()
        chaine_pokemon = rdata_data[offset:offset + bytes_to_read]
        print(chaine_pokemon.hex())

pe.close()
```

```python
bytes_chaine_pokemon = bytearray(chaine_pokemon)
key = bytearray(b"dynamax")

# Xor de la string lue dans .rdata avec la clé
for i in range(0,len(bytes_chaine_pokemon)):
    bytes_chaine_pokemon[i] ^= key[i % 7]

print(bytes_chaine_pokemon)
```

```python
# Décodage de la chaine en base64 avec utf-16 (c'était dur de trouver le bon décodage)
str_chaine = base64.b64decode(bytes_chaine_pokemon).decode('utf-16')
print(str_chaine)
```

```python
output_file_name = "dracaufeu.ps1"

# Écriture de la chaîne dans le fichier
with open(output_file_name, 'w', encoding="utf-16") as file:
    file.write(str_chaine)
```
